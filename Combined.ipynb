```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
df_eng = pd.read_excel("C:/Users/vyache/Downloads/Features_English.xlsx")
df_de = pd.read_excel("C:/Users/vyache/Downloads/Features_German.xlsx")
df_self_reports = pd.read_excel("C:/Users/vyache/Desktop/Questionnaire data_Eng.xlsx")
df_self_reports_ger = pd.read_excel("C:/Users/vyache/Desktop/Questionnaire data_Ger.xlsx")

# Merge the features (df_eng, df_de) with the self-reports (df_self_reports, df_self_reports_ger) on a common identifier (e.g., participant ID)
df_eng_combined = pd.merge(df_eng, df_self_reports, on='Speaker', how='inner')
df_de_combined = pd.merge(df_de, df_self_reports_ger, on='Speaker', how='inner')

# Select the relevant columns: 'F0_sma_de_amean_mean' for Pitch and 'Emotional Resilience'
df_eng_combined = df_eng_combined[['F0_sma_de_skewness_mean', 'Emotional Resilience']]
df_de_combined = df_de_combined[['F0_sma_de_skewness_mean', 'Emotional Resilience']]

# Add a new column to differentiate between English and German datasets
df_eng_combined['Language'] = 'English'
df_de_combined['Language'] = 'German'

# Combine both datasets into one DataFrame
df_combined = pd.concat([df_eng_combined, df_de_combined])

# Scatter Plot for both English and German datasets in one graph
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_combined, x='F0_sma_de_skewness_mean', y='Emotional Resilience', hue='Language', palette={'English': 'blue', 'German': 'green'}, style='Language', markers=['o', 's'])
plt.title('Pitch vs Emotional Resilience (English and German)', fontsize=14)
plt.xlabel('Pitch', fontsize=12)
plt.ylabel('Emotional Resilience (self-reported score)', fontsize=12)
plt.grid(True)
plt.legend(title='Language')
plt.show()
```
